```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
df=pd.read_excel("AdventureWorks.xlsx")

```

```python
df.dtypes
```

```python
df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"])
df.head()
```

```python
round(df["Custo"].sum(),2)
```

```python
#Receita total
round(df["Valor Venda"].sum(),2)
```

```python
df["Lucro total"] = df["Valor Venda"] - df["Custo"]
df.head()
```

```python
lucro_final = round(df["Lucro total"].sum(),2)
```

```python
df["Tempo de Envio"] = (df["Data Envio"]- df["Data Venda"]).dt.days
df.head()
```

```python
media_entrega= df.groupby("Marca")["Tempo de Envio"].mean().reset_index()
```

```python
#Nao posui valores ausentes
df.isnull().sum()
```

```python
lucro_anual_marca = df.groupby([df["Data Venda"].dt.year,"Marca"])["Lucro total"].sum()
lucro_anual_marca
```

```python
pd.options.display.float_format='{:20,.2f}'.format
```

```python
#Produtos mais vendidos em onder Crescente
produtos_vendidos= df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False).reset_index()
```

```python
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="TOTALPRODUTOS VENDIDOS", color="blue")
plt.xlabel("Total")
plt.ylabel("Quantidade")
plt.legend();
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro total"].sum().plot.bar(title="Lucro Anual");
```

```python
df_vendas_2009= df[df["Data Venda"].dt.year == 2009]
df_vendas_2009.head()
```

```python
df_vendas_2009.groupby([df["Data Venda"].dt.month])["Lucro total"].sum().plot(title="Lucro X Mês")
plt.xlabel("mês")
plt.ylabel("Lucro")
plt.legend;
```

```python
#Lucro total das empresas anual de 2009
df_vendas_2009.groupby("Marca")["Lucro total"].sum().plot.bar(title="Lucro 2009 X Marca")

plt.xlabel("marca")
plt.ylabel("lucro total")
plt.xticks(rotation ="horizontal");
```

```python
#Lucro total por CLASSE

df_vendas_2009.groupby("Classe")["Lucro total"].sum().plot.bar(title="Classe X Lucro")
plt.xlabel("Classe")
plt.ylabel("Lucro total")
```
